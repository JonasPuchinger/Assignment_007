```python
# Define functions

# Parse output to get list only consisting of timestamps
def parse_output(output_string):
    parsed_list = []
    output_string = output_string.split("\n")
    for string in output_string:
        if string  is not "":
            parsed_list.append(int(string.split(",")[-1]))
    return parsed_list

# Calculate delta times between keystrokes
def calculate_delta_times(timestamps):
    iterator = 0
    delta_times_list = []
    for time in timestamps:
        if iterator < len(timestamps) - 1:
            delta_times_list.append(timestamps[iterator + 1] - timestamps[iterator])
            iterator += 1
    return delta_times_list
```
